# Scatter Plot Code 

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis, x_axis, 'o', color = "red", label = "Chicago") 
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBar Fare by Month')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
#replace y_axis for y_axis_larger    
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color = 'skyblue', alpha=0.8,
            edgecolors ='black', linewidth= 2 , label='Boston')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.set_xlabel('Fare($)')
ax.set_title('PyBar Fare by Month')
ax.set_xlim(0,50)
ax.legend()
plt.show
```

# Pie Charts 

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       explode = explode,
       colors = colors,
       labels=x_axis,
       autopct= '%.1f%%') 
plt.show()
```
